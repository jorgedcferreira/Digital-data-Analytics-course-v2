# Exploratory analysis | Descriptive analytics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.feature_selection import mutual_info_classif
import seaborn as sns
```

```python
#IMPORT TRAIN DATA TO DATAFRAME
data = pd.read_csv('../data sets/house prices/train.csv', sep= ',', low_memory =False)
data.head()
```

### Análise Univariada


```python
data.groupby(['SaleCondition'])[['SaleCondition']].count().plot.bar()
```

```python
description = data.describe().T
description['stdev'] = data.std()
description['skewness'] = data.skew()
description['kurtosis'] = data.kurt()
description
```

```python
corr =data.corr()
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
```

```python
 sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
